# 🌾 Descifrando cultivos antiguos con Machine Learnig: introducción al análisis de isótopos con Machine Learning

## 1️⃣ ¿Qué son los isótopos?

Los isótopos son como *huellas químicas* que las plantas absorben del suelo y el agua. Analizando granos carbonizados de hace miles de años, podemos inferir el clima, el tipo de suelo o cómo se cultivaban.

Los isótopos son versiones ligeramente diferentes de un mismo elemento químico.

🌱 Las plantas absorben isótopos del suelo y el agua según el clima, el tipo de tierra y cómo se cultivaron.

🔬 Al analizar granos carbonizados se puede inferir inferir:

| Isótopo | Nos dice |
|:---|:---|
| **Carbono (δ13C)** | Si la planta sufrió **sequía** o fue **regada** |
| **Nitrógeno (δ15N)** | Si se usó **estiércol** como fertilizante |

## 4️⃣ Pregunta 1: ¿Se puede datar un cereal por sus isótopos?

**Contexto**: A veces no tenemos carbono-14 para fechar. ¿Los isótopos nos dan pistas?

📊 **Lo que hacemos**:
- Entrenamos un modelo con muestras de **época conocida**
- Le damos solo: δ13C, δ15N, latitud, longitud
- Le pedimos que **adivine el periodo** (Neolítico, Bronce, Hierro)

🤖 **Modelos**: Random Forest, XGBoost, SVM

💡 **Utilidad práctica**: Un *"método alternativo para fechar"* cuando no hay otros datos.

---

## 5️⃣ Pregunta 2: Trigo vs Cebada — ¿podemos distinguirlos?

**Contexto**: Muchos granos arqueológicos están **demasiado quemados** para reconocer su forma.

📊 **Lo que hacemos**:
- Usamos muestras donde **sabemos** si es trigo o cebada
- El modelo aprende patrones isotópicos y geográficos
- Luego lo aplicamos a muestras problema

🤖 **Modelos**: Random Forest, Regresión Logística, XGBoost

💡 **Utilidad práctica**: *"Identificar especie cuando la morfología no es concluyente"*

🎯 **Resultado esperado**: Si trigo y cebada tienen diferentes necesidades hídricas, sus isótopos deberían reflejarlo

---

## 6️⃣ Pregunta 3: ¿Se puede predecir el estrés hídrico (δ13C)?

**Contexto**: El δ13C nos indica si un cultivo sufrió sequía o fue regado.

📊 **Lo que hacemos**:
- Predecimos el **valor numérico** de δ13C
- Usamos: especie, época, ubicación, δ15N
- El modelo aprende relaciones complejas

🤖 **Modelos**: Random Forest Regressor, Regresión Lineal, XGBoost Regressor

💡 **Utilidad práctica**: *"Estimar si un cultivo antiguo creció en condiciones secas o húmedas"*

📈 **Ejemplo**: δ13C muy negativo → más estrés hídrico → posible sequía

---

## 7️⃣ Pregunta 4: ¿Podemos adivinar el país de origen solo con isótopos?

**Contexto**: Diferentes países tienen diferentes climas, suelos y tradiciones agrícolas. ¿Se refleja esto en los isótopos?

📊 **Lo que hacemos**:
- El modelo recibe: δ13C, δ15N, latitud, longitud
- Debe predecir: **país** (Francia, España, Italia, Reino Unido, Alemania...)
- Es un problema de **clasificación multiclase**

🤖 **Modelos**: Random Forest, XGBoost, SVM

💡 **Utilidad práctica**: *"¿Podríamos saber de dónde vino un grano comercializado en la antigüedad?"* (Este no me convence, pero hay que hacer 5.()

🌍 **Ejemplo**: Un grano encontrado en un barco hundido en el Mediterráneo → ¿era de Egipto, Grecia o Italia?

---

## 8️⃣ Pregunta 5: Clustering — ¿Qué nos dicen los datos sin etiquetas?

**Contexto**: A veces **no sabemos** qué buscar. Dejamos que el algoritmo encuentre **grupos naturales**.

📊 **Lo que hacemos**:
- Agrupamos yacimientos por su **firma isotópica** (δ13C, δ15N)
- Sin decirle al modelo a qué país, periodo o especie pertenecen
- Vemos si los grupos coinciden con **regiones geográficas** o **tradiciones culturales**

🤖 **Modelos**: K-means, DBSCAN

💡 **Utilidad práctica**: *"Descubrir patrones ocultos de circulación de cultivos o técnicas compartidas"*

🗺️ **Visualización**: Mapa con clusters de colores
