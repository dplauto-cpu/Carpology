## 1. Introduction

This project was carried out as part of an assessed practical component of a Machine Learning course. While the focus is not on carpological research, datasets from academic research have been used, and basic guidance has been provided by a member of the Research Group **isoTOPIK·Lab** at the *University of Burgos* (UBU). This is a preliminary study that is expected to be revisited at some point from a purely archaeological perspective.

### 1.1 Archaeological Context
The domestication of cereals (wheat and barley) during the Neolithic revolution transformed human societies. However, identifying charred archaeological seeds to species level is seriously challenging due to degradatation and carbonisation. Stable isotope analysis (δ13C and δ15N) offers an alternative: δ13C indicates water stress, while δ15N reflects aridity or manuring intensity.

### 1.2 Research Question
Is it possible to train a ML model to distinguish barley vs wheat isotopes, and integrate the data into a coherent geographical and chronological archaeological framework?

### 1.3 Objectives
- **Primary:** Build a supervised model to classify barley vs wheat
- **Secondary:** Identify isotopic clusters using unsupervised learning
- **Tertiary:** Create interactive maps for archaeological interpretation

## 3. Archaeological Interpretation and Results

### 3.1 Key Archaeological Findings

#### Finding 1: Isotopes alone cannot perfectly classify barley vs wheat

The ceiling of ~75% accuracy (AUC 0.824) indicates that isotopic signatures of barley and wheat overlap substantially. This is not a model limitation but reflects archaeological reality: the same species grown in different environmental conditions produces different isotopic values.

**Archaeological implication:** Environmental context (water availability, soil type, agricultural practices) must be considered alongside species identification.

#### Finding 2: Water stress (δ13C) is the strongest predictor

With 41% feature importance, δ13C is the primary driver of classification. More negative δ13C values (better water availability) are associated with:
- Western Mediterranean samples
- Neolithic and early Bronze Age periods
- Wheat more than barley (but with overlap)

**Archaeological implication:** Water management (irrigation, field location) may be as important as species choice in ancient agricultural systems.

#### Finding 3: Four distinct isotopic clusters reveal agricultural intensity gradients

The K-means clustering (k=4) reveals a continuum of agricultural conditions:

| Cluster | Ecological interpretation | Dominant region | Chronological peak |
|:---|:---|:---|:---|
| 0 | Optimal conditions (low stress) | Western Mediterranean | Neolithic |
| 1 | Moderate stress | Central Mediterranean | Bronze Age |
| 2 | High stress | Eastern Mediterranean | Iron Age |
| 3 | Extreme aridity / manuring | Eastern Mediterranean | Late Iron Age |

**Archaeological implication:** A clear east-west gradient of increasing aridity is detectable in the isotopic data. This aligns with modern climatic patterns and suggests that ancient farmers adapted their agricultural strategies to local environmental constraints.

#### Finding 4: Chronological trends show increasing water stress over time

Mean δ13C values become progressively more positive (increasing water stress) from Neolithic (-24.0‰) to Iron Age (-21.5‰). This could reflect:
- Climate drying during the Bronze and Iron Ages (well documented in Eastern Mediterranean palaeoclimate records)
- Expansion of agriculture into more marginal areas
- Changes in crop selection or irrigation practices

#### Finding 5: Geography matters as much as chronology

Latitude and longitude together account for 23% of feature importance. The Mediterranean basin classification (Western, Central, Eastern) captures regional agricultural traditions that persist across chronological periods.

**Archaeological implication:** Regional agricultural practices are deeply embedded in local environmental and cultural contexts. A "one-size-fits-all" model of ancient agriculture is inadequate.

### 3.2 Interactive Map

An interactive map was created using **Folium** with a vintage National Geographic base layer (Esri NatGeo_World_Map) to facilitate archaeological interpretation.

**Map features:**

| Feature | Description |
|:---|:---|
| **Base layer** | National Geographic style (sepia/olive tones) for archaeological aesthetic |
| **Markers** | Colour-coded by isotopic cluster (brown, sienna, peru, chocolate) |
| **Marker clustering** | Density-based grouping with custom colours (<10, 10-50, >50 samples) |
| **Popups** | Display site name (`Archaeological_Site`), cereal type, δ13C, δ15N, period, basin |
| **Legend (top-right)** | Cluster colours with mean isotopic values |
| **Legend (bottom-centre)** | Density ranges for sample aggregation |
| **Scale** | Bottom-right corner, repositioned for clarity |

**Interpretation using the map:**
- **Western Mediterranean (Iberia, southern France, Italy west of 11°E):** Dominated by Clusters 0 and 1 (optimal to moderate conditions). Sites like *Castellaro dell'Uscio* (Italy) show high δ13C values consistent with good water availability.
- **Central Mediterranean (Italy east of 11°E, Adriatic, Greece west of 23°E):** Mixed clusters 1 and 2. Transition zone between western humidity and eastern aridity.
- **Eastern Mediterranean (Aegean, Anatolia, Levant):** Dominated by Clusters 2 and 3 (high stress to extreme aridity). Sites like *Case Bastione* show elevated δ15N values suggesting manuring or arid soils.
