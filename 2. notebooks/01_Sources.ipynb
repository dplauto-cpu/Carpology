```python
# Cargamos los dos datasets
carpo_nuevo = pd.read_csv('carpo_nuevo.csv')  # Ajusta la ruta si es necesario
trigo_cebada = pd.read_csv('trigo_cebada_clean_ML.csv', na_values=['', ' ', 'NA'])

print("=== CARPO_NUEVO ===")
print(f"Shape: {carpo_nuevo.shape}")
print(f"Columnas: {carpo_nuevo.columns.tolist()[:10]}...")  # Primeras 10

print("\n=== TRIGO_CEBADA_CLEAN_ML ===")
print(f"Shape: {trigo_cebada.shape}")
print(f"Columnas: {trigo_cebada.columns.tolist()[:10]}...")
```

```python
# ============================================
# UNIÓN APROXIMADA POR COORDENADAS (KDTree)
# ============================================

from scipy.spatial import KDTree
import numpy as np
import pandas as pd

print("=== UNIÓN APROXIMADA: AÑADIENDO SITIO DESDE TRIGO_CEBADA ===\n")

# 1. Preparar trigo_cebada: solo filas con 'site' no nulo
df_trigo_con_sitio = trigo_cebada[trigo_cebada['site'].notna()].copy()
print(f"📊 Filas con 'site' en trigo_cebada: {len(df_trigo_con_sitio)}")

# 2. Extraer coordenadas de trigo_cebada (las que tienen site)
coords_trigo = df_trigo_con_sitio[['latitud', 'longitud']].dropna().values
print(f"📊 Puntos de referencia en trigo_cebada: {len(coords_trigo)}")

if len(coords_trigo) == 0:
    print("❌ No hay puntos de referencia. Verifica que 'latitud' y 'longitud' tengan valores.")
else:
    # 3. Crear KDTree para búsqueda eficiente
    tree = KDTree(coords_trigo)
    
    # 4. Buscar el sitio más cercano para cada punto en carpo_nuevo
    nombres_sitio = []
    distancias = []
    
    for idx, row in carpo_nuevo.iterrows():
        if pd.notna(row['Latitude_N']) and pd.notna(row['Longitude_E']):
            punto = np.array([[row['Latitude_N'], row['Longitude_E']]])
            distancia, idx_match = tree.query(punto, k=1)
            
            # Umbral: 0.2 grados ≈ 22 km (para sitios arqueológicos cercanos)
            umbral = 0.2  # grados decimales (~22 km en el Mediterráneo)
            
            if distancia[0] < umbral:
                sitio_encontrado = df_trigo_con_sitio.iloc[idx_match[0]]['site']
                nombres_sitio.append(sitio_encontrado)
                distancias.append(distancia[0])
            else:
                nombres_sitio.append(None)
                distancias.append(None)
        else:
            nombres_sitio.append(None)
            distancias.append(None)
    
    # 5. Añadir columna al DataFrame
    carpo_nuevo['Site_Name_Trigo'] = nombres_sitio
    carpo_nuevo['distancia_coords'] = distancias
    
    # 6. Verificar resultados
    encontrados = carpo_nuevo['Site_Name_Trigo'].notna().sum()
    print(f"\n✅ Sitios encontrados: {encontrados} de {len(carpo_nuevo)} ({encontrados/len(carpo_nuevo)*100:.1f}%)")
    
    # 7. Mostrar ejemplos
    print("\n🔍 Ejemplos de sitios añadidos (con distancia):")
    ejemplos = carpo_nuevo[carpo_nuevo['Site_Name_Trigo'].notna()][['Entry_ID', 'Site_Name', 'Site_Name_Trigo', 'distancia_coords', 'Latitude_N', 'Longitude_E']].head(10)
    print(ejemplos.to_string())
    
    # 8. (Opcional) Eliminar columna de distancia si no la necesitas
    # carpo_nuevo = carpo_nuevo.drop(columns=['distancia_coords'])

print("\n✅ Unión aproximada completada")
```

```python
## ============================================
# ENFOQUE COMBINADO: coordenadas + merge tradicional
# ============================================

# 1. Añadir Site_Name_Trigo mediante coordenadas (más fiable)
print("\n=== AÑADIENDO SITE_NAME_TRIGO POR COORDENADAS ===\n")

# Redondear coordenadas
carpo_nuevo['lat_round'] = carpo_nuevo['Latitude_N'].round(2)
carpo_nuevo['lon_round'] = carpo_nuevo['Longitude_E'].round(2)
carpo_nuevo['coord_key'] = carpo_nuevo['lat_round'].astype(str) + '_' + carpo_nuevo['lon_round'].astype(str)

trigo_cebada['lat_round'] = trigo_cebada['latitud'].round(2)
trigo_cebada['lon_round'] = trigo_cebada['longitud'].round(2)
trigo_cebada['coord_key'] = trigo_cebada['lat_round'].astype(str) + '_' + trigo_cebada['lon_round'].astype(str)

# Crear mapeo
df_sitios_trigo = trigo_cebada[['coord_key', 'site']].drop_duplicates(subset=['coord_key'])
df_sitios_trigo = df_sitios_trigo[df_sitios_trigo['site'].notna()]
print(f"📊 Mapeo desde trigo_cebada: {len(df_sitios_trigo)} claves únicas")

# Añadir columna
carpo_nuevo['Site_Name_Trigo'] = carpo_nuevo['coord_key'].map(df_sitios_trigo.set_index('coord_key')['site'])

encontrados = carpo_nuevo['Site_Name_Trigo'].notna().sum()
print(f"✅ Sitios encontrados: {encontrados} de {len(carpo_nuevo)} ({encontrados/len(carpo_nuevo)*100:.1f}%)")

# Limpiar columnas temporales
carpo_nuevo = carpo_nuevo.drop(columns=['lat_round', 'lon_round', 'coord_key'])
trigo_cebada = trigo_cebada.drop(columns=['lat_round', 'lon_round', 'coord_key'])

# 2. Ahora sí, hacer el merge tradicional para el resto de columnas
print("\n=== MERGE TRADICIONAL PARA EL RESTO DE COLUMNAS ===\n")

# Renombrar trigo_cebada
trigo_cebada_renombrado = trigo_cebada.rename(columns={
    'cereal': 'Genus',
    'taxon': 'Species',
    'country': 'Modern_Country',
    # 'site' ya no la renombramos porque ya la añadimos
    'Cultural period': 'Cultural_Horizon',
    'latitud': 'Latitude_N',
    'longitud': 'Longitude_E',
    'd13C': 'IRMS_d13C_Collagen',
    'd15N': 'd15N_Collagen'
})

# Columnas clave para el merge (sin Site_Name_Trigo)
columnas_clave = ['Genus', 'Species', 'Modern_Country', 
                  'Cultural_Horizon', 'Latitude_N', 'Longitude_E', 
                  'IRMS_d13C_Collagen', 'd15N_Collagen']

# Columnas extra
columnas_extra = [col for col in trigo_cebada_renombrado.columns 
                  if col not in carpo_nuevo.columns and col not in columnas_clave]

print(f"Columnas a añadir: {columnas_extra}")

# Hacer merge
neo_plants = pd.merge(
    carpo_nuevo,
    trigo_cebada_renombrado[columnas_clave + columnas_extra],
    on=columnas_clave,
    how='left'
)

print(f"\n✅ Merge completado. Shape: {neo_plants.shape}")
print(f"Columnas finales: {neo_plants.columns.tolist()}")
```

```python
# ============================================
# LIMPIAR Y UNIFICAR NOMBRES DE SITIO
# ============================================

print("\n=== LIMPIEZA DE COLUMNAS DE SITIO ===\n")

# 1. Eliminar columnas basura que no necesitamos
columnas_a_eliminar = ['distancia_coords', 'site', 'cronologia', '%C', '%N', 'C:N', 
                       'peso_mg', 'pre-treatment', 'contexto', 'method']

for col in columnas_a_eliminar:
    if col in carpo_nuevo.columns:
        carpo_nuevo = carpo_nuevo.drop(columns=[col])
        print(f"  Eliminada: {col}")

# 2. Renombrar Site_Name_Trigo para claridad (opcional)
if 'Site_Name_Trigo' in carpo_nuevo.columns:
    carpo_nuevo.rename(columns={'Site_Name_Trigo': 'Site_Name_Trigo'}, inplace=True)

# 3. Crear columna unificada 'Archaeological_Site'
#    Prioridad: primero Site_Name (MAIA), luego Site_Name_Trigo (trigo_cebada)
carpo_nuevo['Archaeological_Site'] = carpo_nuevo['Site_Name'].fillna(carpo_nuevo['Site_Name_Trigo'])

# 4. Limpiar valores vacíos o 'nan' en texto
carpo_nuevo['Archaeological_Site'] = carpo_nuevo['Archaeological_Site'].replace('', pd.NA)
carpo_nuevo['Archaeological_Site'] = carpo_nuevo['Archaeological_Site'].replace('nan', pd.NA)

# 5. Verificar cobertura final
cobertura_maia = carpo_nuevo['Site_Name'].notna().sum()
cobertura_trigo = carpo_nuevo['Site_Name_Trigo'].notna().sum()
cobertura_final = carpo_nuevo['Archaeological_Site'].notna().sum()

print(f"\n📊 COBERTURA FINAL DE NOMBRES DE SITIO:")
print(f"   • Desde MAIA (Site_Name):      {cobertura_maia} de {len(carpo_nuevo)} ({cobertura_maia/len(carpo_nuevo)*100:.1f}%)")
print(f"   • Desde trigo_cebada (site):   {cobertura_trigo} de {len(carpo_nuevo)} ({cobertura_trigo/len(carpo_nuevo)*100:.1f}%)")
print(f"   • Combinada (Archaeological_Site): {cobertura_final} de {len(carpo_nuevo)} ({cobertura_final/len(carpo_nuevo)*100:.1f}%)")

# 6. Mostrar ejemplos
print("\n🔍 EJEMPLOS DE NOMBRES UNIFICADOS:")
ejemplos = carpo_nuevo[carpo_nuevo['Archaeological_Site'].notna()][
    ['Entry_ID', 'Archaeological_Site', 'Site_Name', 'Site_Name_Trigo', 'Latitude_N', 'Longitude_E']
].head(15)
print(ejemplos.to_string())

# 7. Opcional: eliminar columnas originales de sitio (si no las necesitas)
# carpo_nuevo = carpo_nuevo.drop(columns=['Site_Name', 'Site_Name_Trigo'])

print("\n✅ Limpieza completada. Dataset listo para continuar con el merge.")
```

```python
# ============================================
# GUARDAR EL DATASET CON ARCHAELOGICAL_SITE
# ============================================

# Guardar como CSV (sin sobrescribir el original)
carpo_nuevo.to_csv('carpo_nuevo_con_sitios.csv', index=False)
print("✅ Dataset guardado como 'carpo_nuevo_con_sitios.csv'")
print(f"Dimensiones: {carpo_nuevo.shape}")
print(f"Columnas: {carpo_nuevo.columns.tolist()}")
```
